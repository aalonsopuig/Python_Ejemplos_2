## File Management

### How Python Manages Files

**Opening Files**
To work with files in Python, use the open() function:
file = open('filename.txt', 'mode', encoding='utf-8')

Modes:
- 'r' → Read mode (for reading, file must exist)
- 'w' → Write mode (creates or overwrites the file)
- 'a' → Append mode (adds new content to the end of the file if it exists)

**Why use utf-8?**
Using encoding='utf-8' allows Python to handle special characters correctly, such as accented letters or symbols like ñ, é, €.

**New Line Character \n**
The character \n is used to indicate a line break when writing to a file. Example:

    file.write("Hello\n")

**Writing to a File** To write content to a file:

    file.write("Some text\n")

This writes the text and then moves to the next line.

**Reading from a File** To read all lines from a file:

    lines = file.readlines()

This returns a list of strings, where each string represents a line from the file.

**Why use strip()?**
The strip() method removes any extra spaces and the newline character (\n) from the end of each line.

    print(line.strip())

**Closing a File** Always close the file after reading or writing to free system resources and save changes:

    file.close()

**Using with** (Recommended Syntax) Using with automatically closes the file when finished:

    with open('file.txt', 'r', encoding='utf-8') as file:
       lines = file.readlines()

**Checking if a File Exists** To check if a file exists before trying to read it:

    import os
    if os.path.exists('file.txt'):
       print("File found.")


### Basic File Writing

This program demonstrates how to create a text file and write three lines of text into it using Python. It opens a file in write mode, adds content, and then closes the file properly.

```python
# Open the file in write mode ('w' creates a new file or overwrites if it already exists)
file = open('my_file.txt', 'w')  # Open the file named "my_file.txt" in write mode

# Write three lines to the file
file.write('First line\n')       # Write the first line with a newline character
file.write('Second line\n')      # Write the second line with a newline character
file.write('Third line\n')       # Write the third line with a newline character

# Close the file manually
file.close()                     # Close the file to ensure data is written and file is saved

print('Three lines have been written to the file "my_file.txt".')  # Inform the user of the successful write
```

### Basic File Writing and Reading with Exception Handling

This program demonstrates how to write three lines of text to a file and then read and display them on the screen. It includes basic exception handling to ensure that errors during file operations are managed appropriately.

```python
# Program: Basic File Writing and Reading with Exception Handling

# Name of the file to use
file_name = 'my_file.txt'  # Define the name of the file as "my_file.txt"

# Attempt to open the file and write to it
try:
    # Open the file in write mode ('w' means write mode)
    file = open(file_name, 'w', encoding='utf-8')           # Open file for writing with UTF-8 encoding
    # Write three lines to the file
    file.write('First line\n')                              # Write the first line with a newline
    file.write('Second line\n')                             # Write the second line with a newline
    file.write('Third line\n')                              # Write the third line with a newline
except IOError:                                             # Handle any I/O errors that occur while writing
    print("An error occurred while writing to the file.")   # Inform the user about the error
finally:
    # Ensure the file is closed
    file.close()                                            # Close the file even if an error occurred

# Attempt to open the file and read from it
try:
    # Open the file in read mode ('r' means read mode)
    file = open(file_name, 'r', encoding='utf-8')           # Open the file for reading with UTF-8 encoding
    # Read all lines from the file
    lines = file.readlines()                                # Read all lines into a list
except IOError:                                             # Handle any I/O errors that occur while reading
    print("An error occurred while reading the file.")      # Inform the user about the error
finally:
    # Ensure the file is closed
    file.close()                                            # Close the file after reading

# Print the lines read from the file
print("File content:")                                      # Print a header for the file content
for line in lines:                                          # Iterate over each line in the list
    print(line.strip())                                     # Print the line without trailing newline characters
```

### Interactive File Storage

This program allows the user to add items to a text file interactively. Each item is appended to the file. The process ends when the user presses ENTER without typing anything. Afterward, the program reads the file and displays all the stored items.

```python
# Program: Interactive File Storage (Ends with ENTER)

# Ask the user to provide the file name
file_name = input("Enter file name: ")                                  # Get the name of the file from the user

# Ask the user to input items to store in the file
try:
    # Open the file in append mode to store items
    # If the file does not exist, it creates it; if it exists, it adds data at the end
    file = open(file_name, 'a', encoding='utf-8')                       # Open file for appending in UTF-8 encoding
    
    while True:  # Loop to accept multiple inputs
        item = input("Enter an item to add (press ENTER to finish): ")  # Ask for a new item or ENTER to end
        if item == "":                                                  # If the input is empty, stop adding items
            break                                                       # Exit the loop
        file.write(item + '\n')                                         # Write the item followed by a newline to the file
        print(f"'{item}' has been added.")                              # Confirm the item has been added
except IOError:                                                         # Catch file writing errors
    print("An error occurred while writing to the file.")               # Display error message if writing fails
finally:
    file.close()                                                        # Always close the file after writing, even if an error occurred

# Now read the file and display all stored items
try:
    # Open the file in read mode to retrieve the items
    file = open(file_name, 'r', encoding='utf-8')                       # Open the file for reading
    print("\nItems:")                                                   # Print a header for the list of items
    for item in file:                                                   # Iterate through each line in the file
        print(item.strip())                                             # Print the item without the newline character
except IOError:                                                         # Catch file reading errors
    print("An error occurred while reading the file.")                  # Display error message if reading fails
finally:
    file.close()                                                        # Always close the file after reading, even if an error occurred
```

### Guest List Management Program (with file saving)

This program is an interactive Guest List Management System that allows the user to manage a list of guest names. It includes features to:

- Add guests,

- List all guests with numbering,

- Remove a guest by number,

- Clear the entire list, and

- Save/load the guest list from a fixed text file to preserve data between sessions.

When the program starts, it loads the guest list from a file if it exists. When the user exits, the list is saved automatically to the file. The program uses a simple text-based menu to interact with the user.

```python
import os  # Import module to check if the file exists

# Define the file name where the guest list will be stored
FILE_NAME = "guest_list.txt"                                                # Fixed name for the file used to load/save the guest list

# Create an empty list to store the names of the guests
guest_list = []                                                             # Global variable to hold guest names

def load_guest_list():
    """Loads guest names from file if it exists."""
    if os.path.exists(FILE_NAME):                                           # Check if the file exists
        with open(FILE_NAME, 'r', encoding='utf-8') as file:                # Open the file for reading
            for line in file:                                               # Loop through each line in the file
                name = line.strip()                                         # Remove trailing newline characters
                if name != "":                                              # Make sure the line is not empty
                    guest_list.append(name)                                 # Add the name to the guest list

def save_guest_list():
    """Saves the current guest list to file."""
    with open(FILE_NAME, 'w', encoding='utf-8') as file:                    # Open the file for writing
        for guest in guest_list:                                            # Loop through all guest names
            file.write(guest + '\n')                                        # Write each name on a new line

def display_menu():
    """Displays the main menu and asks for the user’s choice."""
    print("\nMenu:")                                                        # Print menu header
    print("A: Add - Add a new guest.")                                      # Menu option to add guests
    print("L: List - Show the guest list.")                                 # Menu option to list all guests
    print("R: Remove - Remove a guest from the list.")                      # Option to remove guest
    print("C: Clear - Clear the entire guest list.")                        # Option to clear the list
    print("E: Exit - Exit the program and save the guest list.")            # Option to exit and save
    return input("Enter an option (A, L, R, C, E): ").upper()               # Get user input and convert to uppercase

def add_guest():
    """Adds guest names to the list until the user presses ENTER."""
    while True:                                                             # Start a loop to allow multiple guest entries
        name = input("Enter the guest's name (press ENTER to finish): ")    # Prompt for guest name
        if name == "":                                                      # If the input is empty, stop adding
            break                                                           # Exit the loop
        guest_list.append(name)                                             # Add the guest name to the list
        print(name, "has been added to the guest list.")                    # Confirm addition

def list_guests():
    """Displays all guests with numbering."""
    print("\nGuest list:")                                                  # Header
    if len(guest_list) == 0:                                                # Check if list is empty
        print("The list is empty.")                                         # Notify if empty
        return                                                              # Exit function
    sorted_list = sorted(guest_list)                                        # Sort the list alphabetically
    index = 1                                                               # Start numbering from 1
    for name in sorted_list:                                                # Loop through each guest
        print(index, ":", name)                                             # Print numbered name
        index += 1                                                          # Increase index for next guest

def remove_guest():
    """Removes a guest by their position in the list."""
    if len(guest_list) == 0:                                                # If the list is empty
        print("The list is empty.")                                         # Show message
        return                                                              # Exit function
    list_guests()                                                           # Show current guests
    try:
        index_to_remove = int(input("Enter the number of the guest to remove: ")) - 1  # Ask for number to remove
        if 0 <= index_to_remove < len(guest_list):                          # Check if index is valid
            removed = guest_list.pop(index_to_remove)                       # Remove the guest
            print(removed, "has been removed from the list.")               # Confirm deletion
        else:
            print("Invalid number, please enter a valid number.")           # Handle out-of-range input
    except ValueError:                                                      # If the input is not a number
        print("Please enter a valid integer.")                              # Show error message

def clear_guest_list():
    """Clears all guests from the list after confirmation."""
    confirmation = input("Are you sure you want to clear the entire list? (Y/N): ").upper()  # Ask for confirmation
    if confirmation == "Y":                                                 # If confirmed
        guest_list.clear()                                                  # Empty the list
        print("The guest list has been cleared.")                           # Notify user
    else:
        print("Clear operation cancelled.")                                 # Notify cancellation

# Load guest list from file when the program starts
load_guest_list()                                                           # Attempt to load previous guest list

# Main program loop
while True:
    option = display_menu()                                                 # Show menu and get user input
    if option == "A":
        add_guest()                                                         # Call function to add guest
    elif option == "L":
        list_guests()                                                       # Call function to list guests
    elif option == "R":
        remove_guest()                                                      # Call function to remove a guest
    elif option == "C":
        clear_guest_list()                                                  # Call function to clear the list
    elif option == "E":
        save_guest_list()                                                   # Save the list before exiting
        print("Guest list saved. End of the program.")                      # Confirm save and exit
        break                                                               # Exit the loop
    else:
        print("Invalid option, please try again.")                          # Handle invalid input
```

### Blood Pressure Visualization Program

This program uses the matplotlib library to plot systolic and diastolic blood pressure readings. The data sequence illustrates the progression of blood pressure from a normal state to mild hypertension, followed by a drop to slightly low values and then recovery to normal levels. This example demonstrates how changes in blood pressure can be effectively visualized using line charts.

```python
import matplotlib.pyplot as plt  # Import the module used for plotting

# Blood pressure data adjusted to show a realistic progression
systolic_pressures = [120, 125, 130, 135, 140, 135, 130, 125, 120, 115]     # Systolic values over time
diastolic_pressures = [80, 82, 85, 88, 90, 85, 80, 78, 75, 73]              # Diastolic values over time

# Initialize the figure for the plot and set its size
plt.figure(figsize=(10, 5))                                                 # Create a plot with width 10 and height 5

# Plot the systolic pressures with red color, circles for points, and lines connecting them
plt.plot(systolic_pressures, 'r-o', label='Systolic')                       # Red line with markers for systolic pressure

# Plot the diastolic pressures with blue color, circles for points, and lines connecting them
plt.plot(diastolic_pressures, 'b-o', label='Diastolic')                     # Blue line with markers for diastolic pressure

# Add a title to the graph
plt.title('Blood Pressure Record')                                          # Title of the chart

# Label the X-axis
plt.xlabel('Measurement Number')                                            # X-axis label indicating the measurement order

# Label the Y-axis
plt.ylabel('Blood Pressure (mmHg)')                                         # Y-axis label indicating pressure values in mmHg

# Add a legend to differentiate between the two lines
plt.legend()                                                                # Display legend to identify systolic and diastolic lines

# Enable the grid to make the chart easier to read
plt.grid(True)                                                              # Show a grid in the background of the chart

# Display the final chart
plt.show()                                                                  # Render the chart on screen
```

### Blood Pressure Recording Program

This program allows the user to manually enter a patient's systolic and diastolic blood pressure readings. The values are stored in a text file, and once data collection is complete, the results can be visualized in a line graph using Matplotlib. This is useful for tracking and observing blood pressure trends over time.

Pseudocode:

Program:

```python
import matplotlib.pyplot as plt                                                         # Import module for plotting

# File name where blood pressure readings will be saved
file_name = 'blood_pressure_readings.txt'                                               # Define fixed file name

# Input phase: Add blood pressure readings and store in file
try:
    file = open(file_name, 'a', encoding='utf-8')                                       # Open file in append mode (create if not exists)

    while True:
        # Ask user for systolic pressure (ENTER to stop)
        systolic_input = input("Enter systolic pressure (press ENTER to finish): ")
        if systolic_input == "":
            break                                                                       # Stop input when ENTER is pressed

        try:
            systolic = int(systolic_input)                                              # Convert systolic input to integer
        except ValueError:
            print("Invalid input. Please enter a number for systolic pressure.")        # Error if not a number
            continue                                                                    # Ask again

        try:
            diastolic_input = input("Enter diastolic pressure: ")                       # Ask for diastolic pressure
            diastolic = int(diastolic_input)                                            # Convert to integer
        except ValueError:
            print("Invalid input. Please enter a number for diastolic pressure.")       # Error if not a number
            continue                                                                    # Ask again

        # Save the valid values in the file as systolic/diastolic
        file.write(f"{systolic}/{diastolic}\n")                                         # Store the reading in the file
        print("Pressure reading saved.")                                                # Confirm to user

except IOError:
    print("An error occurred while writing to the file.")                               # File write error
finally:
    file.close()                                                                        # Always close the file

# Prepare to read values from the file
systolic_pressures = []                                                                 # List to store systolic values
diastolic_pressures = []                                                                # List to store diastolic values

try:
    file = open(file_name, 'r', encoding='utf-8')                                       # Open file for reading

    for line in file:
        clean_line = line.strip()                                                       # Remove newline characters
        parts = clean_line.split('/')                                                   # Split line into systolic and diastolic parts
        systolic_pressures.append(int(parts[0]))                                        # Add systolic value to list
        diastolic_pressures.append(int(parts[1]))                                       # Add diastolic value to list

except IOError:
    print("An error occurred while reading the file.")                                  # File read error
finally:
    file.close()                                                                        # Always close the file

# Plot the blood pressure data if there are values
if systolic_pressures and diastolic_pressures:
    plt.figure(figsize=(10, 5))                                                         # Set the plot size
    plt.plot(systolic_pressures, 'r-o', label='Systolic')                               # Plot systolic in red
    plt.plot(diastolic_pressures, 'b-o', label='Diastolic')                             # Plot diastolic in blue
    plt.title('Blood Pressure Record')                                                  # Title of the graph
    plt.xlabel('Measurement Number')                                                    # Label for x-axis
    plt.ylabel('Blood Pressure (mmHg)')                                                 # Label for y-axis
    plt.legend()                                                                        # Show legend to identify lines
    plt.grid(True)                                                                      # Show background grid
    plt.show()                                                                          # Display the plot
else:
    print("No data to display.")                                                        # Message if there are no readings
```

### Blood Glucose Monitoring Program


This program is designed to help monitor a patient's daily blood glucose levels. It allows the user to input glucose readings (in mg/dL), stores them in a list during runtime, and saves the data to a text file for future sessions. Upon startup, the program automatically loads previously saved values from the file if it exists.

The program provides a menu with the following functionalities:

* **Add** a new glucose value.
* **List** all recorded values.
* **Graph** the trend of glucose levels using a line chart.
* **Exit** the program and save all collected data.

The chart includes two horizontal lines representing the lower and upper normal glucose limits (70 and 100 mg/dL), helping to visually identify abnormal values. This tool provides a simple yet practical way to track glucose trends over time, combining basic data storage, user interaction, and graphical visualization using `matplotlib`.


```python
# Blood Glucose Monitoring Program

import matplotlib.pyplot as plt                             # Import plotting module

# File where glucose values will be stored
file_name = 'glucose_readings.txt'                          # Fixed file name

# Load existing data from file into the glucose list
glucose_values = []                                         # Initialize the list to store glucose values

try:
    file = open(file_name, 'r', encoding='utf-8')           # Try to open the file in read mode
    for line in file:
        line = line.strip()                                 # Remove newline characters
        if line != "":
            glucose_values.append(int(line))                # Add each value to the list
    file.close()                                            # Close the file after reading
except FileNotFoundError:
    pass                                                    # If the file doesn't exist yet, we start with an empty list

# Function to add a new glucose value
def add_glucose():
    value = input("Enter today's glucose value (mg/dL): ")  # Ask for input
    if value == "":
        print("No value entered.")
        return
    if not value.isdigit():
        print("Invalid input. Please enter a number.")
        return
    glucose_values.append(int(value))                       # Add to list
    print("Glucose value saved.")

# Function to list all stored glucose values
def list_glucose():
    if not glucose_values:
        print("No values recorded yet.")
        return
    print("Registered glucose values (mg/dL):")
    for i, value in enumerate(glucose_values, start=1):
        print("Measurement", i, "→", value)

# Function to show the glucose chart
def show_chart():
    if not glucose_values:
        print("No data to display.")
        return
    plt.figure(figsize=(10, 5))                             # Set plot size
    plt.plot(glucose_values, 'bo-', label='Glucose')        # Line chart in blue
    plt.axhline(70, color='orange', linestyle='--', label='Lower normal limit (70 mg/dL)')  # Lower limit
    plt.axhline(100, color='green', linestyle='--', label='Upper normal limit (100 mg/dL)')  # Upper limit
    plt.title('Blood Glucose Over Time')                    # Chart title
    plt.xlabel('Measurement Number')                        # X-axis label
    plt.ylabel('Glucose (mg/dL)')                           # Y-axis label
    plt.legend()                                            # Show legend
    plt.grid(True)                                          # Show grid
    plt.show()                                              # Display the chart

# Function to save data to file
def save_to_file():
    try:
        file = open(file_name, 'w', encoding='utf-8')       # Open file in write mode
        for value in glucose_values:
            file.write(str(value) + '\n')                   # Write each value to a new line
        file.close()                                        # Close file
        print("Data saved to file.")
    except IOError:
        print("An error occurred while saving to file.")

# Main loop
while True:
    print("\nMenu:")
    print("A - Add today's glucose value")
    print("L - List all values")
    print("G - Show chart")
    print("E - Exit and save")
    option = input("Choose an option (A, L, G, E): ").upper()

    if option == "A":
        add_glucose()
    elif option == "L":
        list_glucose()
    elif option == "G":
        show_chart()
    elif option == "E":
        save_to_file()
        print("Exiting program.")
        break
    else:
        print("Invalid option. Please try again.")
```
